Задача проекта: создать модель, которая будет предсказывать стоимость автомобиля по его характеристикам.
Есть тестовый датасет test.csv, в котором нет колонки стоимости автомобиля. Для автомобилей в этом датасете нужно предсказать их стоимость на основе обученной модели. 

Готового тренировочного датасета для построения модели нет. Поэтому нужно спарсить с сайта auto.ru данные по соответствующим в тестовом датасете признакам и значения цены авто в рублях в отношении примерно 80 - 100 тысяч автомобилей. Для этого интересует только один регион - Москва / Московская область. Теоретически можно брать другие российские регионы, но там цена может отличаться до 5 % в сравнении с Москвой.

```python
# ниже в этой ячейке изложены функции изъятия данных, они нормально работали, когда я удачно изымал информацию
# и записывал её в переменную soup 

# а вот код парсинга из следующей за ней ячейки поработал два дня и перестал работать

# функция выявления цены авто

def get_price(soup):
    
    try:    
        item_0 = soup.text.split(sep="рублей")
        item_0_0 = []
        for i in item_0:
            if i.__contains__("по цене"): 
                item_0_0.append(i)
                try:
                    item_0_0 = item_0_0[0].split(sep="по цене ")       
                    item_0_0 = item_0_0[1].split(sep="\xa0") 
                    length = len(item_0_0)
                    if length == 2:
                        item_0_0 = item_0_0[0] + item_0_0[1]
                    elif length == 3:
                        item_0_0 = item_0_0[0] + item_0_0[1] + item_0_0[2]
                    elif length == 4:
                        item_0_0 = item_0_0[0] + item_0_0[1] + item_0_0[2] + item_0_0[3]   
                    elif length == 5:
                        item_0_0 = item_0_0[0] + item_0_0[1] + item_0_0[2] + item_0_0[3] + item_0_0[4]  
                    else:
                        item_0_0 = 0
                    item_0_0 = int(item_0_0)
                                    
                except LookupError:
                    print("В строке нет слов 'по цене'")   
    except LookupError:
        print("В строке нет слов 'рублей'") 
    return item_0_0          

# код выявления признаков марки, модели и происхождения авто (Европа или Япония)
                    
def get_model(soup):
    
    try:    
        item_0_1 = soup.find('h1',class_="CardHead__title").get_text() # название модели
        item_0_1_0 = item_0_1.split()[0] # это - только марка авто
        item_0_1_0 = item_0_1_0.upper() 
        item_0_1_1 = item_0_1.split()[1:] # здесь определяем модель авто
        
        if item_0_1_1[-1] == 'Рестайлинг':
            item_0_1_1 = item_0_1_1[:-1]
        length = len(item_0_1_1)
        if length == 1:
            item_0_1_1 = item_0_1_1[0]
        elif length == 2:
            item_0_1_1 = item_0_1_1[0] + "_" + item_0_1_1[1] 
        elif length == 3:
            item_0_1_1 = item_0_1_1[0] + "_" + item_0_1_1[1] + "_" + item_0_1_1[2] 
        elif length == 4:
            item_0_1_1 = item_0_1_1[0] + "_" + item_0_1_1[1] + "_" + item_0_1_1[2] + "_" + item_0_1_1[3]   
        elif length == 5:
            item_0_1_1 = item_0_1_1[0] + "_" + item_0_1_1[1] + "_" + item_0_1_1[2] + "_" + item_0_1_1[3] + "_" + item_0_1_1[4]  

        if item_0_1_0 == 'BMW' or item_0_1_0 == 'VOLKSWAGEN' or item_0_1_0 ==  'MERCEDES'\
        or item_0_1_0 == 'AUDI'  or  item_0_1_0 == 'SKODA' or  item_0_1_0 == 'VOLVO': 
            item_16_16 = 'EUROPEAN'
        else:
            item_16_16 = 'JAPANESE'
       
    except LookupError:
        print("Неверный запрос")   
    
    return item_0_1_0, item_0_1_1, item_16_16               

# код выявления года выпуска авто
                    
def get_productionDate(soup):
    
    try:    
        item_1 = soup.find('li',class_="CardInfoRow CardInfoRow_year").get_text()
        item_1_1 = item_1[-4:]
        item_1_1 = int(item_1_1)
                     
    except LookupError:
        print("Неверный запрос")   
    
    return item_1_1  

# код выявления пробега
                    
def get_mileage(soup):
    
    try:    
        item_2 = soup.find('li',class_="CardInfoRow CardInfoRow_kmAge").get_text()
        item_2_2 = item_2[6:].split(sep="\xa0")
        del item_2_2[-1]
        length = len(item_2_2)
        if length == 2:
            item_2_2 = item_2_2[0] + item_2_2[1] 
        elif length == 3:
            item_2_2 = item_2_2[0] + item_2_2[1] + item_2_2[2] 
        elif length == 4:
            item_2_2 = item_2_2[0] + item_2_2[1] + item_2_2[2] + item_2_2[3]   
        item_2_2 = int(item_2_2)   
        
    except LookupError:
        print("Неверный запрос")   
    
    return item_2_2   

# код выявления типа кузова

def get_bodytype(soup):
    
    try:    
        item_3 = soup.find('li',class_="CardInfoRow CardInfoRow_bodytype").get_text() 
        item_3_3 = item_3[5:]
        
    except LookupError:
        print("Неверный запрос")   
    
    return item_3_3   

# код выявления цвета

def get_color(soup):
    
    try:    
        item_4 = soup.find('li',class_="CardInfoRow CardInfoRow_color").get_text()
        item_4_4 = item_4[4:]
        
    except LookupError:
        print("Неверный запрос")   
    
    return item_4_4   

# код выявления характеристик двигателя: литраж (если имеется), мощность и тип топлива 

def get_engine_description(soup):
    
    try:    
        item_5 = soup.find('li',class_="CardInfoRow CardInfoRow_engine").get_text()
        item_5_5 = item_5[9:].split(sep="/")
        if len(item_5_5) > 2:
            item_5_5_1 = item_5_5[0]
            item_5_5_1 = float(item_5_5_1[:-3])
            item_5_5_2 = item_5_5[1]
            item_5_5_2 = int(item_5_5_2[1:-6])
            item_6 = item_5_5[2]
            item_6 = item_6[1:]
        else:
            item_5_5_1 = 0.0
            item_5_5_2 = item_5_5[0]
            item_5_5_2 = int(item_5_5_2[1:-6])
            item_6 = item_5_5[1]
            item_6 = item_6[1:]
        
    except LookupError:
        print("Неверный запрос")   
    
    return item_5_5_1,item_5_5_2,item_6   

# код выявления коробки передач

def get_transmission(soup):
    
    try:    
        item_7 = soup.find('li',class_="CardInfoRow CardInfoRow_transmission").get_text()        
        item_7_7 = item_7[7:]

    except LookupError:
        print("Неверный запрос")   
    
    return item_7_7   

# код выявления привода

def get_wd(soup):
    
    try:    
        item_8 = soup.find('li',class_="CardInfoRow CardInfoRow_drive").get_text()        
        item_8_8 = item_8[6:]

    except LookupError:
        print("Неверный запрос")   
    
    return item_8_8   

# код выявления руля

def get_wheel(soup):
    
    try:    
        item_9 = soup.find('li',class_="CardInfoRow CardInfoRow_wheel").get_text()       
        item_9_9 = item_9[4:]

    except LookupError:
        print("Неверный запрос")   
    
    return item_9_9   

# код выявления состояния авто

def get_state(soup):
    
    try:    
        item_10 = soup.find('li',class_="CardInfoRow CardInfoRow_state").get_text()      
        item_10_10 = item_10[9:]

    except LookupError:
        print("Неверный запрос")   
    
    return item_10_10   

# код выявления количества владельцев

def get_ownersCount(soup):
    
    try:    
        item_11 = soup.find('li',class_="CardInfoRow CardInfoRow_ownersCount").get_text()     
        item_11_11 = item_11[9:].split(sep="\xa0") # потом еще доработаем в зависимости от данных в этом столбце

    except LookupError:
        print("Неверный запрос")   
    
    return item_11_11   

# код выявления информации о ПТС

def get_PTS(soup):
    
    try:    
        item_12 = soup.find('li',class_="CardInfoRow CardInfoRow_pts").get_text()   
        item_12_12 = item_12[3:]
        
    except LookupError:
        print("Неверный запрос")   
    
    return item_12_12   

# код выявления информации о сроке владения последним владельцем

def get_owningTime(soup):
    
    try:    
        item_13 = soup.find('li',class_="CardInfoRow CardInfoRow_owningTime").get_text()   
        item_13_13 = item_13[8:]# потом еще доработаем в зависимости от данных в этом столбце
        
    except LookupError:
        print("Неверный запрос")   
    
    return item_13_13   

# код выявления информации о том, было ли растаможено авто

def get_customs(soup):
    
    try:    
        item_14 = soup.find('li',class_="CardInfoRow CardInfoRow_customs").get_text()   
        item_14_14 = item_14[7:]
        
    except LookupError:
        print("Неверный запрос")   
    
    return item_14_14 

# код выявления информации о комплектации

def get_customs(soup):
    
    try:     
        item_15_list = soup.find_all('div',class_="ComplectationGroupsDesktop__row") 
        # - элементы экстерьера (доп. обработка)
        item_15_pure_list = []
        for i in item_15_list:
            item_15_pure_list.append(i.get_text()) 
        
        item_15_15 = item_15_pure_list[0].split(sep="•") # Проблема со смешением в строке решена
        
        for i, equip in enumerate(item_15_15):
            if equip.__contains__("Элементы экстерьера"): # есть и другие (скрытые) разделы, нужно доработать
                item_15_15[i] = equip[0:-19] 
        
    except LookupError:
        print("Неверный запрос")   
    
    return item_15_15 
```
